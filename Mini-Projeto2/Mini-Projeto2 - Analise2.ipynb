```python
# Imports
!pip install pyreadline 
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.image as mpimg
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path)
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
```

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise2/brand-vehicleCount.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
sns.set_style("whitegrid")
g = sns.catplot(y="brand", data=df, kind="count", height=7)
g.ax.set_title('Veículos por Marca')
g.ax.set_xlabel('Número de Veículos', fontsize=15)
g.ax.set_ylabel('Marca', fontsize=15)
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
```

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise2/vehicletype-gearbox-price.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(13, 7))
sns.set_style("whitegrid")
ax = sns.barplot(x="vehicleType", y="price",hue="gearbox", data=df)
ax.set_title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio", fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_ylabel('Preço Médio', fontsize=15)
```
