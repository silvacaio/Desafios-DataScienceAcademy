```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path)
```

# Distribuição de Veículos com base no Ano de Registro

### Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise1/vehicle-distribution.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(9, 5))
ax = sns.distplot(df['yearOfRegistration'])

ax.set_xlabel('Ano de registro', fontsize=15)
ax.set_ylabel('Densidade (KDE)', fontsize=15)
```

# Variação da faixa de preço pelo tipo de veículo

### Crie um Boxplot para avaliar os outliers

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise1/price-vehicleType-boxplot.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(9, 5))
ax = sns.boxplot(x='vehicleType', y='price', data=df)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_ylabel('Range de Preço', fontsize=15)
```

# Contagem total de veículos à venda conforme o tipo de veículo

### Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise1/count-vehicleType.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(12, 7))
ax = sns.countplot(x='vehicleType', data=df)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_ylabel('Total de Veículos para Venda', fontsize=15)

vehicleTypes = df['vehicleType'].value_counts().values
qtdVehicleTypes = range(len(vehicleTypes))

for p in ax.patches:
        ax.annotate(p.get_height(), (p.get_x()+0.2, p.get_height()+60))
```
