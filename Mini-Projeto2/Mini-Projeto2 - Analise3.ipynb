```python
# Imports
!pip install pyreadline 
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.image as mpimg
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path)
```

```python
print(df.columns.tolist())
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise3/fueltype-vehicleType-price.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(13, 7))
ax = sns.barplot(x="fuelType", y="price",hue="gearbox", data=df)
ax.set_title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio", fontsize=15)
ax.set_xlabel('Tipo de Combustível', fontsize=15)
ax.set_ylabel('Preço Médio', fontsize=15)
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
```

```python
#Original
plt.figure(figsize=(13, 7))
img = mpimg.imread('plots/Analise3/vehicletype-fueltype-power.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
# Desenvolvimento e réplica do desafio solicitado
plt.figure(figsize=(13, 7))
ax = sns.barplot(x="vehicleType", y="powerPS",hue="gearbox", data=df)
ax.set_title("Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio", fontsize=15)
ax.set_xlabel('Tipo de Veículo', fontsize=15)
ax.set_ylabel('Preço Médio', fontsize=15)
```
